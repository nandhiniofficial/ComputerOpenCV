```python
import cv2 
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt 
```

```python
vid  = cv2.VideoCapture(0)
re,f = vid.read()
re
```

```python
while(True):
    ret, frame = vid.read()
    video = cv2.cvtColor(frame,cv2.COLOR_RGB2BGR)  
  

    cv2.imshow('Video',video)
    
    if cv2.waitKey(30) & 0xFF == ord('q'):
        break
        
        
vid.release()
cv2.destroyAllWindows()
```

```python
kernel =q np.ones((5,5),np.uint8)
```

```python
erosion = cv2.erode(video,kernel,iterations = 1)
```

```python
kernel
```

```python
dilation = cv2.dilate(video,kernel,iterations = 2)
```

```python
open = cv2.morphologyEx(video,cv2.MORPH_OPEN,kernel)
```

```python
close  = cv2.morphologyEx(video,cv2.MORPH_CLOSE,kernel)
```

```python
gradient = cv2.morphologyEx(video, cv2.MORPH_GRADIENT, kernel)
```

```python
tophat = cv2.morphologyEx(video, cv2.MORPH_TOPHAT, kernel)
```

```python
blackhat = cv2.morphologyEx(video, cv2.MORPH_BLACKHAT, kernel)
```

```python
titles = ["org_image","erosion","dilation","opening","closing","gradient","tophat","blackhat"]
videos = [video,erosion,dilation,open,close,gradient,tophat,blackhat]

for i in range(len(videos)):
    plt.subplot(2,4,i+1)
    plt.imshow(videos[i])
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
    
plt.show()
```
